#Question

numpy array

```python
#Create a 1D NumPy array with values from 0 to 9.
import numpy as np
arr = np.arange(10)
print(arr)
```

```python
#Create a 2D NumPy array of shape (3, 3) with values from 0 to 8.
arr = np.arange(9).reshape(3, 3)
print(arr)
```

```python
#Create a NumPy array with all elements as 1 of size 5.
arr = np.ones(5)
print(arr)
```

```python
#Create a NumPy array with all elements as 0 of shape (2, 3).
arr = np.zeros((2, 3))
print(arr)
```

```python
#Create a NumPy array of 10 random numbers between 0 and 1.
arr = np.random.rand(10) #only +ve value but randn() give both +ve and _ve vale
print(arr)
```

```python
#Create a 1D NumPy array with values from 10 to 50 (both inclusive).
arr = np.arange(10, 51)
print(arr)
```

```python
#Find the shape of a given NumPy array.
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(arr.shape)
```

```python
#Create a 2D identity matrix of size 4.
arr = np.eye(3,3)
print(arr)
```

```python
#Find the data type of elements in a given NumPy array.
arr = np.array([1, 2, 3.0])
print(arr.dtype)
```

```python
#Compute the sum of all elements in a NumPy array.
arr = np.array([1, 2, 3, 4, 5])
print(np.sum(arr))
```

```python
#Create a 3x3 matrix with values ranging from 0 to 8 and then transpose it.
arr = np.arange(9).reshape(3, 3)
transposed = np.transpose(arr)
print(transposed)
```

```python
#Create a NumPy array with 20 random numbers and find the mean and standard deviation.
arr = np.random.randn(20)
print(arr)
mean  = np.mean(arr)
std_dev = np.std(arr)
print("mean :", mean, "std :",std_dev)
```

```python
#Create a 1D NumPy array with values from 0 to 9 and replace all odd numbers with -1
arr = np.arange(10)
arr[arr % 2 != 0] = -1
print(arr)
```

```python
#Create a 2D array of shape (4, 5) with random integers from 1 to 20 and find the maximum element in each row.
arr = np.random.randint(1,21, size=(4,5))
max_num = np.max(arr,axis=0)
print(max_num)
```

```python
lst1 = [1,2,3]
lst2 = [4,5,6]
lst3 = [7,8,9]
arr =np.array([lst1, lst2, lst3])
max_num0 = np.max(arr, axis=1)
max_num1 = np.max(arr, axis=0)
print(max_num0, max_num1)
```

```python
#Stack two given 1D arrays vertically and horizontally.
arr0 = np.array([1,2,3])
arr1 = np.array([4,5,6])
vstack = np.vstack((arr0, arr1))
hstack = np.hstack((arr0, arr1))
print("vstack : ", vstack)
print()
print("hstack : ", hstack)
```

```python
#Create a 2D array of shape (5, 5) with random values and normalize it (i.e., scale the values to be between 0 and 1).
arr = np.random.rand(5,5)
arr_min = arr.min()
arr_max = arr.max()
normalized = (arr - arr_min) / (arr_max - arr_min)
print(normalized)
```

```python
#Create a 1D array of 10 random integers and sort them in ascending order.
arr = np.random.randint(1, 100, size=10)
sorted_arr = np.sort(arr)
reversed_arr = sorted_arr[::-1]
print("original arr : ",arr)
print("sorted_arr : ",sorted_arr)
print("reversed_arr : ",reversed_arr)
```
